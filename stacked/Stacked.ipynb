```python
from IPython.display import display, Javascript, HTML
from random import random
import numpy as np
import json
```

```python
n = 5 # number of series
m = 58 # number of values per series
height = 400
width = 700
```

```python
# Returns an array of m psuedorandom, smoothly-varying non-negative numbers.
# Inspired by Lee Byronâ€™s test data generator.
# http://leebyron.com/streamgraph/
def bumps(m):
    values = []

    # Initialize with uniform random values in [0.1, 0.2).
    values = [0.1 + 0.1 * random() for _ in range(m)]

    # Add five random bumps.
    for _ in range(5):
        x = 1 / (0.1 + random())
        y = 2 * random() - 0.5
        z = 10 / (0.1 + random())
        for i in range(m):
            w = (i / m - y) * z
            values[i] += x * np.exp(-w * w);

    # Ensure all values are positive.
    for i in range(m):
        values[i] = max(0, values[i]);

    return values

data = [bumps(m) for _ in range(n)]
```

```python
display(Javascript("require.config({paths: {d3: 'd3.v5.min'}});"))
display(Javascript(filename="stacked.js"))
display(HTML(filename="stacked.css.html"))
```

```python
def draw_stacked(data, width=500, height=500):
    display(Javascript("""
        (function(element){
            require(['stacked'], function(stacked) {
                stacked(element.get(0), %s, %d, %d);
            });
        })(element);
    """ % (json.dumps(data), width, height)))
```

```python
draw_stacked(data, width=width, height=height)
```
